## Feature engineering

Ajout d'une variable de proportion des graisses saturées

```python
tx_sat = data['saturated-fat_100g']/data['fat_100g']
data['taux_sat_fat'] = tx_sat.values

# Fonction qui supprime les valeurs hors échelle
def hors_echelle(valeur):
    if valeur < 0 or valeur > 1:
        return 0
    else:
        return valeur
data['taux_sat_fat'] = [hors_echelle(val) for val in data['taux_sat_fat']]
```

```python
col = 'taux_sat_fat'
print("minimum = ", data['taux_sat_fat'].min())
print("maximum = ", data['taux_sat_fat'].max())        
print("moyenne = ", data['taux_sat_fat'].mean())
print("médiane = ", data['taux_sat_fat'].median())
print("mode = ", data['taux_sat_fat'].mode())
print("variance = ",data['taux_sat_fat'].var(ddof=0))
print("ecart-type = ",data['taux_sat_fat'].std(ddof=0))
print("skewness = ",data['taux_sat_fat'].skew())
print("kurtosis = ",data['taux_sat_fat'].kurtosis())
data['taux_sat_fat'].hist(bins = 30)
plt.show()
data.boxplot(column='taux_sat_fat', vert=False)
plt.show()
```

```python
#prédiction Score
data["score_pred"] = model.params[0] + data["fat_100g"] * model.params[1] + data["carbohydrates_100g"] * model.params[2] + data["energy_100g"] * model.params[3] + data["saturated_fat_100g"] * model.params[4] + data["sugars_100g"] * model.params[5] 
```

```python
data["diff_score"] = abs(data["score_pred"] - data["nutrition_score_fr_100g"])
print("minimum = ", data["diff_score"].min())
print("maximum = ", data["diff_score"].max())
print("moyenne = ", data["diff_score"].mean())
print("médiane = ", data["diff_score"].median())
print("mode = ", data["diff_score"].mode())
print("variance = ",data["diff_score"].var(ddof=0))
print("ecart-type = ",data["diff_score"].std(ddof=0))
print("skewness = ",data["diff_score"].skew())
print("kurtosis = ",data["diff_score"].kurtosis())
data["diff_score"].hist(bins = 100)
plt.show()
plt.boxplot(data["diff_score"], vert=False)
plt.show()
plt.boxplot(data["diff_score"], vert=False, showfliers=False)
plt.show()
```

```python
col_mandat = ['energy_100g', 'fat_100g']

data_temp = data.dropna(axis='index', subset=col_mandat, inplace=False)


X = data_temp[['fat_100g']]
Y = data_temp['energy_100g']
X = X.copy() # On modifiera X, on en crée donc une copie
X['intercept'] = 1.
result = sm.OLS(Y, X).fit() # OLS = Ordinary Least Square (Moindres Carrés Ordinaire)
a,b = result.params['fat_100g'],result.params['intercept']
print (a,b)
str_eq = "y = " + str(round(a,2)) + " x + " + str(round(b,0))
```

```python
plt.figure(1, figsize=(9, 6))
plt.plot(data.fat_100g,data.energy_100g, "o")
plt.plot(np.arange(100),[a*x+b for x in np.arange(100)],label=str_eq)
plt.ylabel("Energie pour 100g")
plt.xlabel("Matière grasse pour 100g")
plt.title("Relation Matières Grasses / Energie")
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols("energy_100g ~ fat_100g + carbohydrates_100g + sugars_100g ", data=data).fit()
print(model.params)
print(model.summary())
```

```python
#prédiction énergie
data["energy_pred"] = model.params[0] + data["fat_100g"] * model.params[1] + data["carbohydrates_100g"] * model.params[2] + data["sugars_100g"] * model.params[3]
```

```python
data["diff_energy"] = (data["energy_pred"] - data["energy_100g"])/data["energy_pred"]
print("minimum = ", data["diff_energy"].min())
print("maximum = ", data["diff_energy"].max())
print("moyenne = ", data["diff_energy"].mean())
print("médiane = ", data["diff_energy"].median())
print("mode = ", data["diff_energy"].mode())
print("variance = ",data["diff_energy"].var(ddof=0))
print("ecart-type = ",data["diff_energy"].std(ddof=0))
print("skewness = ",data["diff_energy"].skew())
print("kurtosis = ",data["diff_energy"].kurtosis())
data["diff_energy"].hist(bins = 100)
plt.show()
plt.boxplot(data["diff_energy"], vert=False, showfliers=False)
plt.show()
```
